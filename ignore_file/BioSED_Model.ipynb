```python
import os
import pandas as pd
from tqdm import tqdm
from birdnetlib import Recording
from birdnetlib.analyzer import Analyzer

# 1. Inicializar el analizador de BirdNET (descarga el modelo automáticamente)
print("Cargando modelo de BirdNET...")
analyzer = Analyzer()

# 2. Configurar las rutas de tus datos
DATA_DIR = "dataset_aves"  # Cambia esto por tu ruta real
resultados = []

# 3. Recorrer las carpetas de las 10 especies
especies_carpetas = [f for f in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, f))]

print("Iniciando el procesamiento de audios...")
for especie_carpeta in especies_carpetas:
    ruta_carpeta = os.path.join(DATA_DIR, especie_carpeta)
    archivos_audio = [f for f in os.listdir(ruta_carpeta) if f.endswith(('.wav', '.mp3', '.flac'))]
    
    print(f"\nProcesando especie: {especie_carpeta} ({len(archivos_audio)} archivos)")
    
    for archivo in tqdm(archivos_audio):
        ruta_audio = os.path.join(ruta_carpeta, archivo)
        
        try:
            # 4. Pasar el audio a BirdNET
            # min_conf=0.10 para capturar cantos tenues; luego podemos filtrar más estricto
            recording = Recording(analyzer, ruta_audio, min_conf=0.10)
            recording.analyze()
            
            # 5. Extraer las detecciones encontradas en el audio
            for deteccion in recording.detections:
                resultados.append({
                    "archivo_original": archivo,
                    "especie_carpeta": especie_carpeta,  # La etiqueta que tú le diste originalmente
                    "ave_detectada_comun": deteccion["common_name"],
                    "ave_detectada_cientifico": deteccion["scientific_name"],
                    "inicio_segundo": deteccion["start_time"],
                    "fin_segundo": deteccion["end_time"],
                    "confianza_birdnet": deteccion["confidence"]
                })
                
        except Exception as e:
            print(f"Error procesando {archivo}: {e}")

# 6. Convertir a un DataFrame de Pandas
df_etiquetas_fuertes = pd.DataFrame(resultados)
print(f"\n¡Proceso completado! Se encontraron {len(df_etiquetas_fuertes)} segmentos de canto.")
```

# Entrenamiento de Modelo BioSED (Bioacoustic Sound Event Detection)

Este cuaderno entrena una red neuronal recurrente convolucional (CRNN) para detectar simultáneamente hasta 10 especies de aves y predecir sus probabilidades acústicas a nivel de ventana de tiempo (frame-level).

## Cell 1: Configuración del entorno y Librerías

Importamos los paquetes importantes para procesamiento de señales y redes neuronales, y configuramos una semilla determinística (Seed) para garantizar reproducibilidad en las divisiones de las bases de datos y pesos iniciales de la red.

```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torchaudio
import torchaudio.transforms as T
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score, confusion_matrix, ConfusionMatrixDisplay

# Configuración determinista
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Utilizando dispositivo de cómputo: {device}")
```

## Cell 2: Función de Preprocesamiento de Audio

Convierte archivos en bruto a un Espectrograma Log-Mel 2D estandarizado para la arquitectura CNN, que es el formato acústico estándar para tareas como BioSED.

```python
def preprocess_audio(file_path, target_sr=32000, n_fft=1024, hop_length=320, n_mels=64):
    """
    Carga, resamplea y transforma un archivo de audio WAV/MP3 
    en un Log-Mel Spectrogram 2D tensorial.
    """
    try:
        waveform, sr = torchaudio.load(file_path)
    except Exception as e:
        # Prevención ante errores IO (e.g., simulación o fallos de códec)
        waveform = torch.zeros((1, target_sr * 3))
        sr = target_sr
        
    # Reducimos canales a mono en caso de capturas multicanal o estéreo
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)
        
    # Cambio al rate requerido (32kHz) de forma uniforme
    if sr != target_sr:
        resampler = T.Resample(orig_freq=sr, new_freq=target_sr)
        waveform = resampler(waveform)
        
    mel_transformer = T.MelSpectrogram(
        sample_rate=target_sr,
        n_fft=n_fft,
        hop_length=hop_length,
        n_mels=n_mels,
        power=2.0 # Espectro de potencia antes de la conversión a decibeles
    )
    
    mel_spec = mel_transformer(waveform)
    
    # Transformar a escala logarítmica (Log-Mel)
    amp_to_db = torchaudio.transforms.AmplitudeToDB(stype="power")
    log_mel_spec = amp_to_db(mel_spec)
    
    return log_mel_spec
```

## Cell 3: Dataset PyTorch y DataLoader con Stratificación
Construimos un `DataLoader` para manejar las dimensiones correctas e interpolar padding asumiendo un máximo predefinido de celdas temporales (features visuales).

```python
class BioacousticDataset(Dataset):
    def __init__(self, file_paths, labels, class_to_idx, max_frames=500):
        self.file_paths = file_paths
        self.labels = labels
        self.class_to_idx = class_to_idx
        self.max_frames = max_frames
        self.num_classes = len(class_to_idx)
        
    def __len__(self):
        return len(self.file_paths)
        
    def __getitem__(self, idx):
        path = self.file_paths[idx]
        label_idx = self.class_to_idx[self.labels[idx]]
        
        # Shape: [1, n_mels, time_frames]
        log_mel = preprocess_audio(path)
        
        # Ajustar los frames al límite "max_frames" del batch limitando temporalidad
        n_mels, time_frames = log_mel.shape[1], log_mel.shape[2]
        
        if time_frames < self.max_frames:
            padding = torch.zeros((1, n_mels, self.max_frames - time_frames))
            log_mel = torch.cat((log_mel, padding), dim=2)
        elif time_frames > self.max_frames:
            log_mel = log_mel[:, :, :self.max_frames]
            
        # Target [Time_frames, 10_Species]: El objetivo requiere predecir a nivel frame.
        # Por simplicidad asiganmos la clase detectada global a cada frame (anotación débil).
        target = torch.zeros((self.max_frames, self.num_classes), dtype=torch.float32)
        target[:, label_idx] = 1.0
        
        return log_mel, target

# Indexación del directorio (usando tu estructura de carpetas)
dataset_dir = "dataset_aves"
species_list = sorted([d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))])
class_to_idx = {species: i for i, species in enumerate(species_list)}

paths, ground_truth = [], []
for species in species_list:
    species_dir = os.path.join(dataset_dir, species)
    for fname in os.listdir(species_dir):
        if fname.endswith(('.wav', '.mp3', '.flac')):
            paths.append(os.path.join(species_dir, fname))
            ground_truth.append(species)

# Fallback en caso de que dataset_aves aún no tenga audios
if not paths:
    print("No se encontraron audios reales, usando array dummy limitante.")
    species_list = [f"Specie_{i}" for i in range(10)]
    class_to_idx = {s: i for i, s in enumerate(species_list)}
    for s in species_list:
        paths.extend([f"dummy_{s}_{i}.wav" for i in range(50)])
        ground_truth.extend([s for _ in range(50)])

# Dividimos en Train, Val y Test (70% - 15% - 15%) usando stratify
X_temp, X_test, y_temp, y_test = train_test_split(paths, ground_truth, test_size=0.15, stratify=ground_truth, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=42)

# Loaders
train_dl = DataLoader(BioacousticDataset(X_train, y_train, class_to_idx), batch_size=16, shuffle=True)
val_dl = DataLoader(BioacousticDataset(X_val, y_val, class_to_idx), batch_size=16, shuffle=False)
test_dl = DataLoader(BioacousticDataset(X_test, y_test, class_to_idx), batch_size=16, shuffle=False)

print(f"Data balanceada. Muestras Train: {len(X_train)} | Val: {len(X_val)} | Test: {len(X_test)}")
```

## Cell 4: Arquitectura del Modelo CRNN
Combinamos una extracción de frecuencia en bloques CNN bidimensionales, la colapsamos en *Time-Dimension* usando *Pooling* Adaptativo, y finalmente secuenciamos temporalmente usando un BiGRU con *Multi-Label Head*.

```python
class BioSED_CRNN(nn.Module):
    def __init__(self, num_classes=10, cnn_channels=[16, 32, 64], gru_hidden=128):
        super(BioSED_CRNN, self).__init__()
        
        # 1. 2D CNN Block
        # Entrada: [Batch, 1, Mel, Time]
        # MaxPool reduce estrictamente en la dimensión Mel (frecuencia), preservando el Tiempo.
        self.cnn = nn.Sequential(
            nn.Conv2d(1, cnn_channels[0], kernel_size=3, padding=1),
            nn.BatchNorm2d(cnn_channels[0]),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 1)), # H/2 (Mel), W igual (Time)
            
            nn.Conv2d(cnn_channels[0], cnn_channels[1], kernel_size=3, padding=1),
            nn.BatchNorm2d(cnn_channels[1]),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 1)),
            
            nn.Conv2d(cnn_channels[1], cnn_channels[2], kernel_size=3, padding=1),
            nn.BatchNorm2d(cnn_channels[2]),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 1))
        )
        
        # 2. Pooling de frecuencia para colapsar las Mels restantes
        # Transforma el tensor CNN a dimensión estricta [Batch, C, 1, Time]
        self.freq_pool = nn.AdaptiveAvgPool2d((1, None))
        
        # 3. Bloque temporal Bidireccional GRU
        self.gru = nn.GRU(input_size=cnn_channels[-1], hidden_size=gru_hidden, 
                          num_layers=2, batch_first=True, bidirectional=True)
        
        # 4. Cabezal Lineal a Probabilidades de N-Especies
        # GRU bidirectional dobla la característica saliente (hidden * 2)
        self.fc = nn.Linear(gru_hidden * 2, num_classes)
        
    def forward(self, x):
        # x shape inicial: [B, 1, n_mels, T]
        x = self.cnn(x)              # [B, 64, n_mels_reducida, T]
        x = self.freq_pool(x)        # [B, 64, 1, T] 
        x = x.squeeze(2)             # Removemos dimensión inactiva: [B, 64, T]
        
        x = x.permute(0, 2, 1)       # Preparación para RNN (Batch_first=True) -> [B, T, Features]
        
        x, _ = self.gru(x)           # Pasa por BiGRU: [B, T, 256]
        
        x = self.fc(x)               # [B, T, Num_classes]
        return torch.sigmoid(x)      # Multi-label probability por frame

model = BioSED_CRNN(num_classes=10).to(device)
print(model)
```

## Cell 5: Función de Loss y Bucle de Entrenamiento
Usamos entropía cruzada binaria (`BCELoss`) dado que en una tarea SED, cada especie y vector por frame puede poseer probabilidad `0.0` o `1.0` en un esquema *multiclase*.

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)

def train_bio_model(model, train_loader, val_loader, epochs=25):
    history = {"train_loss": [], "val_loss": []}
    
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            
            optimizer.zero_grad()
            preds = model(X_batch)
            loss = criterion(preds, y_batch)
            
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * X_batch.size(0)
            
        train_loss /= len(train_loader.dataset)
        
        # Fase de validación
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                preds = model(X_batch)
                loss = criterion(preds, y_batch)
                val_loss += loss.item() * X_batch.size(0)
        
        val_loss /= len(val_loader.dataset)
        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        
        if (epoch+1) % 5 == 0 or epoch == 0:
            print(f"Epoch {epoch+1}/{epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
            
    return history

# Descomentar para ejecutar el run real de entrenamiento
history = train_bio_model(model, train_dl, val_dl, epochs=30)
```

## Cell 6: Post-Procesamiento Acústico y Umbralización
En producción, necesitamos devolver el intervalo cronometrado `[A, B]` en segundos donde se detecta el ave activa basada en predicciones de la red neuronal.

```python
def map_frames_to_seconds(raw_probabilities, threshold=0.5, hop_length=320, sample_rate=32000, species_names=None):
    """
    Convierte las matrices del frame-level back en eventos: [Start, End, Especie]
    """
    frame_dur_seconds = hop_length / sample_rate
    binary_preds = (raw_probabilities >= threshold).astype(int)
    
    time_frames, num_classes = binary_preds.shape
    events_detected = []
    
    for c in range(num_classes):
        frames = binary_preds[:, c]
        active_start = None
        
        for t, bit in enumerate(frames):
            if bit == 1 and active_start is None:
                active_start = t * frame_dur_seconds
            elif bit == 0 and active_start is not None:
                active_end = t * frame_dur_seconds
                spec_label = species_names[c] if species_names else f"Class_{c}"
                events_detected.append([active_start, active_end, spec_label])
                active_start = None
                
        # Caso de corte abrupto al final del audio
        if active_start is not None:
            active_end = time_frames * frame_dur_seconds
            spec_label = species_names[c] if species_names else f"Class_{c}"
            events_detected.append([active_start, active_end, spec_label])
            
    return events_detected

# Prueba teórica y limpieza
sample_probs = np.random.rand(500, 10)
events = map_frames_to_seconds(sample_probs, threshold=0.9, species_names=species_list)
print("Eventos extraídos de la señal:", events[:3])
```

## Cell 7: Evaluación Cuantitativa (mAP y Gráficos)
Evalúa las métricas en todo el Test Loader colapsando las formas de todos los lotes de tensor con el fin de calcular de manera global el **Average Precision**.

```python
def evaluate_sed_mAP(model, test_loader):
    model.eval()
    all_targets = []
    all_predictions = []
    
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(device)
            preds = model(X_batch)
            
            # Aseguramos aplanar dimensiones para Sklearn iterando en frames x num_classes
            all_predictions.append(preds.cpu().numpy().reshape(-1, 10))
            all_targets.append(y_batch.numpy().reshape(-1, 10))
            
    # Si dataset no fue encontrado en simulación, break out.
    if len(all_targets) == 0:
        return
        
    flat_targets = np.concatenate(all_targets, axis=0)
    flat_preds = np.concatenate(all_predictions, axis=0)
    
    aps = []
    for c in range(len(species_list)):
        # Condicionado en el caso que hayan ausencias completas del target de la clase C.
        if np.sum(flat_targets[:, c]) > 0:
            ap_val = average_precision_score(flat_targets[:, c], flat_preds[:, c])
            aps.append(ap_val)
            print(f"[{species_list[c]}] AP: {ap_val:.4f}")
        else:
            aps.append(0.0)
            
    mean_map = np.mean(aps)
    print(f"\n---> mAP Global (Frame-level): {mean_map:.4f} <---")
    
    # Plotea Visual de Confianza para visualmente inspeccionar fallos: Test Individual Dummy
    plt.figure(figsize=(10, 4))
    plt.plot(flat_targets[0:500, 0], label=f"Verdad (Especie 0)", linestyle="dashed", color='orange')
    plt.plot(flat_preds[0:500, 0], label=f"Predicciones (Especie 0)", alpha=0.7, color='green')
    plt.title("Sample Inspección Visual de Probabilidades - BioSED")
    plt.xlabel("Línea de Vida temporal (Time Frames)")
    plt.ylabel("P(Especie Activa)")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

# Descomentar para evaluar el modelo ya entrenado real:
evaluate_sed_mAP(model, test_dl)
```
